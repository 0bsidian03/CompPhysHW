```python
A = np.array([[10, -1, 2, 0],
              [-1, 11, -1, 3],
              [2, -1, 10, -1],
              [0, 3, -1, 8]])
b1 = np.array([6, 25, -11, 15])
b2 = np.array([12, -11, 13, -1])
def gauss_seidel(A, b, x0, tol=1e-5, max_iterations=1000):
    n = len(b)
    x = x0.copy().astype(float)
    x_prev = np.zeros(n)
    iterations = 0
    for k in range(max_iterations):
        x_prev = x.copy()
        for i in range(n):
            sum1 = np.dot(A[i, :i], x[:i])
            sum2 = np.dot(A[i, i+1:], x_prev[i+1:])
            x[i] = (b[i] - sum1 - sum2) / A[i, i]
        if np.linalg.norm(x - x_prev, ord=2) < tol:
            iterations = k + 1
            return x, iterations
    iterations = max_iterations
    return x, iterations
n = len(b1)
x0_b1 = np.zeros(n)  # Initial guess as a zero vector
solution_b1, iterations_b1 = gauss_seidel(A, b1, x0_b1)
print("Solution for b1:", solution_b1)
print("Iterations for b1:", iterations_b1)
n = len(b2)
x0_b2 = np.zeros(n)  # Initial guess as a zero vector
solution_b2, iterations_b2 = gauss_seidel(A, b2, x0_b2)
print("Solution for b2:", solution_b2)
print("Iterations for b2:", iterations_b2)

```

the low number of iteration demonstrates why gauss-seidel is generally considered a very good method

```python
A = np.array([[3, 1, 2],
              [1, 5, 3],
              [2, 3, 7]])
b = np.array([[-1],
              [1],
              [3]])
def gram_schmidt_qr(A):
    m, n = A.shape
    Q = np.zeros((m, n))
    R = np.zeros((n, n))
    for j in range(n):
        v = A[:, j]
        for i in range(j):
            q = Q[:, i]
            R[i, j] = np.dot(q, v)
            v = v - R[i, j] * q
        R[j, j] = np.linalg.norm(v)
        Q[:, j] = v / R[j, j]
    return Q, R

Q, R = gram_schmidt_qr(A)
print("Matrix Q:", Q)
print("Matrix R:", R)
```

```python
def back_substitution(R, y):
    n = len(y)
    x = np.zeros_like(y, dtype=float)
    for i in range(n - 1, -1, -1):
        sum_val = 0
        for j in range(i + 1, n):
            sum_val += R[i, j] * x[j]
        x[i] = (y[i] - sum_val) / R[i, i]
    return x

xqr = back_substitution(R, Q.T@b)
print("Solution x using QR decomposition:", xqr)
```
